```python
'''
Name: Sahil S. Naik
PRN: S20111006
Subject: DSL Lab
Assignment: #9
'''
mat1 = [[1,2],[3,5]] #Matrix = [Row1],[Row2], it's a classic 2x2 Matrix
mat2 = [[5,1],[4,2]]

print("Given matrix 1 is :\n{} {}\n{} {}".format(mat1[0][0],mat1[0][1],mat1[1][0],mat1[1][1]))
print("\nGiven matrix 2 is :\n{} {}\n{} {}\n".format(mat2[0][0],mat2[0][1],mat2[1][0],mat2[1][1]))

#Function to add given Matrix
#new1 stores addition of elements from m row, n column
def mat_add(mtx1, mtx2):
    sum_mat = [],[]
    new1 = 0
    print("Addition of Given Matrix is : ")
    for i in range(2):
        for j in range(2):
            new1 = mat1[i][j]+mat2[i][j]
            sum_mat[i].append(new1)
    print("{} {}\n{} {}".format(sum_mat[0][0],sum_mat[0][1],sum_mat[1][0],sum_mat[1][1]))
mat_add(mat1,mat2)

#Function to subtract given Matrix
#new2 stores subtraction of elements from m row, n column
def mat_sub(mtx1, mtx2):
    sub_mat = [],[]
    new2 = 0
    print("\nSubtraction of Given Matrix is : ")
    for i in range(2):
        for j in range(2):
            new2 = mat1[i][j]-mat2[i][j]
            sub_mat[i].append(new2)
    print("{} {}\n{} {}".format(sub_mat[0][0],sub_mat[0][1],sub_mat[1][0],sub_mat[1][1]))  
mat_sub(mat1,mat2)

#Function to multiply given Matrix
def mat_mul(mtx1,mtx2):
    mul_mat = [[0,0],[0,0]]
    print("\nMultiplication of Given Matrix is : ")
    for i in range(2):
        for j in range(2):
            for k in range(2):
                mul_mat[i][j] += mat1[i][k]*mat2[k][j]
    print("{} {}\n{} {}".format(mul_mat[0][0],mul_mat[0][1],mul_mat[1][0],mul_mat[1][1]))
mat_mul(mat1,mat2)

#Function for Matrix Transpose
def mat_tran(mtx1,mtx2):
    tran_mat1 = [[0,0],[0,0]]
    tran_mat2 = [[0,0],[0,0]]
    print("\nTranspose of Given matrix 1 is :")
    for i in range(2):
        for j in range(2):
            tran_mat1[j][i] = mtx1[i][j]
    print("{} {}\n{} {}".format(tran_mat1[0][0],tran_mat1[0][1],tran_mat1[1][0],tran_mat1[1][1]))
    print("\nTranspose of Given matrix 2 is :")
    for k in range(2):
        for q in range(2):
            tran_mat2[q][k] = mtx2[k][q]
    print("{} {}\n{} {}".format(tran_mat2[0][0],tran_mat2[0][1],tran_mat2[1][0],tran_mat2[1][1]))
mat_tran(mat1,mat2)
```
