```python
#Including necessary packages
from countryinfo import CountryInfo as CI
ext = 0
#Declaration of all continents here
conti = ['ASIA', 'AFRICA', 'EUROPE', 'NORTH AMERICA', 'SOUTH AMERICA', 'ANTARCTICA', 'AUSTRALIA']

#Declaration of all countries in their respective continents here

#1.Africa
afri = ['ALGERIA','ANGOLA','BENIN','BOTSWANA','BURUNDI','CABO','CAMEROON','CHAD','COMOROS','CONGO',
        'DGIBOUTI','EGYPT','ERITREA','GABON','GAMBIA','GHANA','GUINEA','KENYA','LIBREIA','LIBYA','MADAGASCAR',
        'MALAWI','MALI','MOROCCO','NAMIBIA','NIGER','NIGERIA','SOUTH AFRICA','SOUTH SUDAN','SUDAN','TOGO',
        'UGANDA','ZAMBIA','ZIMBABWE']

#2.Europe
eur = ['ALBANIA','ANDORRA','ARMENIA','AUSTRIA','BELARUS','BELGIUM','BOSNIA','BULGARIA','CROATIA','CZECH','DENMARK',
       'ESTONIA','FINLAND','FRANCE','GEORGIA','GERMANY','GREECE','HUNGARY','ICELAND','ITALY','KAZAKHSTAN','LATIVA',
        'LITHUANIA','LUXEMBOURG','MALTA','MOLDOVA','MONACO','MONTENEGRO','NETHERLANDS','NORWAY','POLAND','PORTUGAL',
        'ROMANIA','SAN MARINO','SERBIA','SLAVAKIA','SPAIN','SWEDEN','SWITZERLAND','TURKEY','UKRAINE','UNITED KINGDOM']
#3.Asia
asi = ['AFGHANISTAN','ARMENIA','AZERBAIJAN','BANGLADESH','BHUTAN','BRUNEI','CAMBODIA','CHINA','CYPRUS','GEORGIA',
       'HONG KONG','INDIA','INDONESIA','IRAN','IRAQ','ISRAEL','JAPAN','JORDAN','KAZAKHSTAN','KUWAIT','LAOS',
       'LEBANON','MACAU','MALAYSIA','MALDIVES','MONGOLIA','MYANMAR','NEPAL','NORTH KOREA','OMAN','PAKISTAN',
       'PALESTINE','PHILIPPINES','QATAR','RUSSIA','SAUDI ARABIA','SINGAPORE','SRI LANKA','SYRIA','TAIWAN',
       'THIALAND','TIMOR','TURKEY','UNITED ARAB EMIRATES','UZBEKISTAN','VIETNAM','YEMEN']

#4.North America
nor_am = ['ANTIGUA','BAHAMAS','BELIZE','CANADA','COSTA RICA','CUBA','DOMINICA','EL SALVADOR','GRENADA','GUATEMAL',
          'HAITI','HONDURAS','JAMAICA','MEXICO','NICARAGUA','PANAMA','SAINT LUCIA','TRINIDAD','UNITED STATES OF AMERICA']

#5.South America
suth_am = ['ARGENTINA','BRAZIL','CHILE','COLOMBIA','ECUADOR','GUYANA','PARAGUAY','PERU','SURINAME','URUGUAY','VENEZUELA']

#6.Australia
austr = ['AUSTRALIA','FIJI','KIRIBATI','MARSHALL ISLANDS','NAURU','NEW ZEALAND','PALAU','PAPUA NEW GUINEA','SAMOA','TONGA',
          'TUVALU','VANUATU']

#7.Antarctica
antar = ['ADELIE LAND','ARGENTINE ANTARCTICA','BRITISH ANTARCTIC TERRITORY','PETER I ISLAND','QUEEN MAUD LAND','ROSS DEPENDECNY']
while(ext!=1):
    chc = int(input("1 = Display countries from a Continent.\n2 = Display info about a Country\n3 = Exit.\nEnter your Choice : "))

    if chc == 1:
        get_info = input("Enter name of Continent to show its info : ").upper()
        if get_info in conti:
            print("\nCountries in Asia are:")
            for i in asi:
                print(i, end="  ")
        else:
            print("No such Continent found!\nPlease check spelling.")

    elif chc == 2:
        get_info = input("Enter Country name to show its info : ").capitalize()
        if get_info in austr or eur or asi or antar or suth_am or nor_am or afri:
            fetch = CI(get_info)
            print("\nCapital of {} is {}!".format(get_info,fetch.capital()))
            print("Timezone is {}".format(fetch.timezones()))
            print("Regional ares is {}sq.km.".format(fetch.area()))
            print("{} is also known as ".format(get_info),end = "")
            temp2 = fetch.alt_spellings()
            for k in temp2:
                temp = temp2.index(k)
                if temp != -1:
                      print(k, end=", ")
                else:
                    print(k, end=".")
            
        else:
            print("No such Country found!\nPlease check spelling.")

    elif chc == 3:
        print("\nExiting...")
        ext += 1
        break
        
    else:
        print("Invalid Choice!")
```
