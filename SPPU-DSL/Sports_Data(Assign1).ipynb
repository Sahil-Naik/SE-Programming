```python
import random
grpA_crck = []
grpB_badm = []
grpC_ftbl = []

for loop1 in range(random.randint(30,34)):
    crck = random.randint(1,60)
    if crck in grpA_crck:
        pass
    else:
        grpA_crck.append(crck)
    
for loop2 in range(random.randint(30,34)):
    badm = random.randint(1,60)
    if badm in grpB_badm:
        pass
    else:
        grpB_badm.append(badm)
    
for loop3 in range(random.randint(30,34)):
    ftbl = random.randint(1,60)
    if ftbl in grpC_ftbl:
        pass
    else:
        grpC_ftbl.append(ftbl)
        
print("Roll no. of Student playing Cricket :\n",grpA_crck)    
print("\nRoll no. of Student playing Badminton :\n",grpB_badm) 
print("\nRoll no. of Student playing Football :\n",grpC_ftbl)

# Function to get roll no. of students playing Cricket and Badminton only
def Cricket_and_Badminton():
    print("\nRoll no. of students playing Badminton and Cricket are :")
    for i1 in grpA_crck:
        if i1 in grpB_badm and i1 not in grpC_ftbl:
            print(i1, end=" ")            
Cricket_and_Badminton()

# Function to get roll no. of students playing Cricket or Badminton but not both
def Cricket_or_Badminton_nb():
    tmp1 = [] # List returns students playing Cricket only
    tmp2 = [] # List returns students playing Badminton only
    print("\n\nRoll no. of Students playing either Badminton or Cricket are :")
    for i2 in grpA_crck:
        if i2 not in grpB_badm and i2 not in grpC_ftbl:
            tmp1.append(i2)
    for i3 in grpB_badm:
        if i3 not in grpA_crck and i3 not in grpC_ftbl:
            tmp2.append(i3)
    print(tmp1,tmp2)            
Cricket_or_Badminton_nb()

# Function to get roll no. of students playing neither Cricket nor Badminton
def neither_Cricket_nor_Badminton():
    print("\nRoll no. of Students neither playing Cricket nor Badminton :")
    for i4 in grpC_ftbl:
        if i4 not in grpA_crck and i4 not in grpB_badm:
            print(i4, end=" ")    
neither_Cricket_nor_Badminton()

# Function to get roll no. of students playing Cricket and Football only
def Cricket_and_Football():
    print("\n\nRoll no. of Students playing Cricket and Football only :")
    for i5 in grpA_crck:
        if i5 in grpC_ftbl and i5 not in grpB_badm:
            print(i5, end=" ")
Cricket_and_Football()
```
